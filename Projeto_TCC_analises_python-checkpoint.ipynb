<h3> Importando Bibliotecas! </h3> 


```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

<h3> Abrindo  arquivo com os dados e transformando em dataframe! </h3> 

```python
arquivo = "41073_2020_101_MOESM2_ESM.xlsx"
arquivo_lido = pd.read_excel(arquivo)
df_completo = pd.DataFrame(arquivo_lido)
df_completo_copia = df_completo.copy()
df_completo_copia.head(10)
```

<h3> Data Cleaning! </h3>
<h5> 1º) Selecionar as respostas finais e gold standart (primeira fase) e só final para a segunda fase!

```python
#Seleção tanto pela fase (Random ou Paired) quanto pelo tipo de grupo (1 ou 2)
#   > 1ª fase:
df_random_pubmed = df_completo_copia[(df_completo_copia['Study'] == 'Random') & (df_completo_copia['Group'] == 2)]
df_random_biorxiv = df_completo_copia[(df_completo_copia['Study'] == 'Random') & (df_completo_copia['Group'] == 1)]
#   > 2ª fase: 
df_paired_peer_rev = df_completo_copia[(df_completo_copia['Study'] == 'Paired') & (df_completo_copia['Group'] == 2)]
df_paired_preprints = df_completo_copia[(df_completo_copia['Study'] == 'Paired') & (df_completo_copia['Group'] == 1)]
#   > Gold Standart!
df_random_pubmed_Gold = df_completo_copia[(df_completo_copia['Respondent'] == 'Gold Standard') & (df_completo_copia['Group'] == 2)]
df_random_biorxiv_Gold = df_completo_copia[(df_completo_copia['Respondent'] == 'Gold Standard') & (df_completo_copia['Group'] == 1)]

#Seleção pelas respostas finais!
#   > 1ª fase:
df_random_pubmed_final = df_random_pubmed[df_random_pubmed['Respondent'] == 'FINAL']
df_random_biorxiv_final = df_random_biorxiv[df_random_biorxiv['Respondent'] == 'FINAL']
#   > 2ª fase: 
df_paired_peer_rev_final = df_paired_peer_rev[df_paired_peer_rev['Respondent'] == 'FINAL'] 
df_paired_preprints_final = df_paired_preprints[df_paired_preprints['Respondent'] == 'FINAL']

#Concatenando!
DF_random_pubmed_final_gold = pd.concat([df_random_pubmed_final, df_random_pubmed_Gold], axis = 0)
DF_random_biorxiv_final_gold = pd.concat([df_random_biorxiv_final, df_random_biorxiv_Gold], axis = 0)

#df_random_pubmed_final.head()
#df_random_biorxiv_final.head()
#df_paired_peer_rev_final.head()
#df_paired_preprints_final.head()
#df_random_pubmed_Gold.head()
#df_random_biorxiv_Gold.head()
#df_paired_peer_rev_Gold.head()
#df_paired_preprints_Gold.head()
#DF_random_pubmed_final_gold
#DF_random_biorxiv_final_gold

#Selecionando a parte somente com as respostas finais, sem as caracteríticas extraídas (journal, publisher, JIF, etc...)
#   > 1ª fase:
DF_perguntas_pubmed = DF_random_pubmed_final_gold.iloc[0:, 25:]
DF_perguntas_biorxiv = DF_random_biorxiv_final_gold.iloc[0:, 25:]
#   > 2ª fase: 
DF_perguntas_peer_rev = df_paired_peer_rev_final.iloc[0:, 25:]
DF_perguntas_preprints = df_paired_preprints_final.iloc[0:, 25:]

#DF_perguntas_pubmed.head()
#DF_perguntas_biorxiv.head()
#DF_perguntas_peer_rev.head()
#DF_perguntas_preprints.head()
```

<h5> 2º) Trocado as respostas finais pelos valores associados! <h5

```python
dicionario_replace = {"Yes, and the statement reports a conflict of interest" : 1,
                     'Yes, and the statement reports NO conflict of interest': 1,
                     'No statement is present': 0,
                     'Sample size calculation is reported, with parameters': 1,
                     'Sample size calculation is reported, without parameters': 0.5,
                     'Sample size calculation is not reported': 0,
                     'Yes (in figures)': 1,
                     'Yes (in raw data)': 1,
                     'Yes (both)': 1,
                     'Yes (blinded)': 1,
                     'Yes, but timing is not mentioned (or is more than one year before experiments)': 0,
                     'Yes (includes approval and committee)': 1,
                     'Yes (includes approval but no committee)': 0.5,
                     'Yes (exact)': 1,
                     'Yes (range)': 0.5,
                     'Yes': 1,
                     'Partially': 0.5,
                     'No': 0,
                     'Not applicable': np.nan,
                     'Automated/Not applicable': np.nan}

DF_perguntas_pubmed = DF_perguntas_pubmed.replace(to_replace = dicionario_replace, value = None)
DF_perguntas_biorxiv = DF_perguntas_biorxiv.replace(to_replace = dicionario_replace, value = None)
DF_perguntas_peer_rev = DF_perguntas_peer_rev.replace(to_replace = dicionario_replace, value = None)
DF_perguntas_preprints = DF_perguntas_preprints.replace(to_replace = dicionario_replace, value = None)

#DF_perguntas_pubmed.head()
#DF_perguntas_biorxiv
```

<h5> 3º) Adicionando colunas de JIF 2016 e Score Total! </h5>

```python
#DF_random_pubmed_final_gold.head()
DF_perguntas_pubmed['JIF 2016'] = DF_random_pubmed_final_gold['Impact Factor']

#df_paired_peer_rev_final
DF_perguntas_peer_rev['JIF 2016'] = df_paired_peer_rev_final['Impact Factor']

#DF_perguntas_pubmed.head()
#DF_perguntas_biorxiv.head()
#DF_perguntas_peer_rev.head()
#DF_perguntas_preprints.head()
```

<h5> Retirando perguntas que não entram no cálculo de score total! </h5>

```python
DF_perguntas_pubmed.drop(['''Do the title and abstract provide a clear idea of the article's main findings'''], axis = 1, inplace = True)
DF_perguntas_pubmed.drop(['Species_vert'], axis = 1, inplace = True)
DF_perguntas_pubmed.drop(['Species_inv'], axis = 1, inplace = True)
DF_perguntas_pubmed.drop(['Were animals obtained from wild conditions, or bred in a laboratory or other controlled environment_vert'], axis = 1, inplace = True)
DF_perguntas_pubmed.drop(['Were animals obtained from wild conditions, or bred in a laboratory or other controlled environment_inv'], axis = 1, inplace = True)
DF_perguntas_pubmed.drop(['Was the required information easy to find and extract from the article'], axis = 1, inplace = True)
DF_perguntas_pubmed.drop(['%Agreement'], axis = 1, inplace = True)

DF_perguntas_biorxiv.drop(['''Do the title and abstract provide a clear idea of the article's main findings'''], axis = 1, inplace = True) 
DF_perguntas_biorxiv.drop(['Species_vert'], axis = 1, inplace = True)
DF_perguntas_biorxiv.drop(['Species_inv'], axis = 1, inplace = True)
DF_perguntas_biorxiv.drop(['Were animals obtained from wild conditions, or bred in a laboratory or other controlled environment_vert'], axis = 1, inplace = True)
DF_perguntas_biorxiv.drop(['Were animals obtained from wild conditions, or bred in a laboratory or other controlled environment_inv'], axis = 1, inplace = True)
DF_perguntas_biorxiv.drop(['Was the required information easy to find and extract from the article'], axis = 1, inplace = True)
DF_perguntas_biorxiv.drop(['%Agreement'], axis = 1, inplace = True)


DF_perguntas_peer_rev.drop(['''Do the title and abstract provide a clear idea of the article's main findings'''], axis = 1, inplace = True) 
DF_perguntas_peer_rev.drop(['Species_vert'], axis = 1, inplace = True)
DF_perguntas_peer_rev.drop(['Species_inv'], axis = 1, inplace = True)
DF_perguntas_peer_rev.drop(['Were animals obtained from wild conditions, or bred in a laboratory or other controlled environment_vert'], axis = 1, inplace = True)
DF_perguntas_peer_rev.drop(['Were animals obtained from wild conditions, or bred in a laboratory or other controlled environment_inv'], axis = 1, inplace = True)
DF_perguntas_peer_rev.drop(['Was the required information easy to find and extract from the article'], axis = 1, inplace = True)
DF_perguntas_peer_rev.drop(['%Agreement'], axis = 1, inplace = True)

DF_perguntas_preprints.drop(['''Do the title and abstract provide a clear idea of the article's main findings'''], axis = 1, inplace = True) 
DF_perguntas_preprints.drop(['Species_vert'], axis = 1, inplace = True)
DF_perguntas_preprints.drop(['Species_inv'], axis = 1, inplace = True)
DF_perguntas_preprints.drop(['Were animals obtained from wild conditions, or bred in a laboratory or other controlled environment_vert'], axis = 1, inplace = True)
DF_perguntas_preprints.drop(['Were animals obtained from wild conditions, or bred in a laboratory or other controlled environment_inv'], axis = 1, inplace = True)
DF_perguntas_preprints.drop(['Was the required information easy to find and extract from the article'], axis = 1, inplace = True)
DF_perguntas_preprints.drop(['%Agreement'], axis = 1, inplace = True)

#DF_perguntas_random.to_excel('DF_random_respostas.xlsx')
#DF_perguntas_paired.to_excel('DF_paired_respostas.xlsx')
```

<h3> Cálculo dos Scores Totais </h3>

```python
#Somatório das pontuações de cada linha, não considerando os valores NaN
somatorio_pubmed = DF_perguntas_pubmed.iloc[0:, 1:57].sum(axis = 1, skipna = True)
somatorio_biorxiv = DF_perguntas_biorxiv.iloc[0:, 1:57].sum(axis = 1, skipna = True)
somatorio_peer_rev = DF_perguntas_peer_rev.iloc[0:, 1:57].sum(axis = 1, skipna = True)
somatorio_preprint = DF_perguntas_preprints.iloc[0:, 1:57].sum(axis = 1, skipna = True)
```

```python
#Contagem de valores NaN por linha
DF_perguntas_pubmed['# NaN values in row'] = DF_perguntas_pubmed.isnull().sum(axis=1)
DF_perguntas_biorxiv['# NaN values in row'] = DF_perguntas_biorxiv.isnull().sum(axis=1)
DF_perguntas_peer_rev['# NaN values in row'] = DF_perguntas_peer_rev.isnull().sum(axis=1)
DF_perguntas_preprints['# NaN values in row'] = DF_perguntas_preprints.isnull().sum(axis=1)

#DF_perguntas_peer_rev.head()
#DF_perguntas_biorxiv.head()
#DF_perguntas_peer_rev.head()
#DF_perguntas_preprints.head()
```

```python
#Subtração do número total de perguntas com pontuação pela contagem de NaN
DF_perguntas_pubmed['Total number of app questions'] = 57 - DF_perguntas_pubmed['# NaN values in row']
DF_perguntas_biorxiv['Total number of app questions'] = 57 - DF_perguntas_biorxiv['# NaN values in row']
DF_perguntas_peer_rev['Total number of app questions'] = 57 - DF_perguntas_peer_rev['# NaN values in row']
DF_perguntas_preprints['Total number of app questions'] = 57 - DF_perguntas_preprints['# NaN values in row']

#DF_perguntas_preprints.head()
#DF_perguntas_biorxiv.head()
#DF_perguntas_peer_rev.head()
#DF_perguntas_preprints.head()
```

```python
DF_perguntas_pubmed['Score total'] = somatorio_pubmed / DF_perguntas_pubmed['Total number of app questions']
DF_perguntas_biorxiv['Score total'] = somatorio_biorxiv / DF_perguntas_biorxiv['Total number of app questions'] 
DF_perguntas_peer_rev['Score total'] = somatorio_peer_rev / DF_perguntas_peer_rev['Total number of app questions']
DF_perguntas_preprints['Score total'] = somatorio_preprint / DF_perguntas_preprints['Total number of app questions']

#somatorio_pubmed 
#somatorio_biorxiv 
#somatorio_peer_rev 
#somatorio_preprint 
```

```python
#DF_perguntas_pubmed.head()
#DF_perguntas_biorxiv.head()
#DF_perguntas_peer_rev.head()
#DF_perguntas_preprints.head()
```

```python
DF_perguntas_pubmed['JIF 2016'] = DF_random_pubmed_final_gold['Impact Factor']
DF_perguntas_peer_rev['JIF 2016'] = df_paired_peer_rev_final['Impact Factor']
#DF_perguntas_peer_rev
```

```python
DF_perguntas_pubmed_copy = DF_perguntas_pubmed.copy()
DF_perguntas_biorxiv_copy = DF_perguntas_biorxiv.copy()
DF_perguntas_peer_rev_copy = DF_perguntas_peer_rev.copy()
DF_perguntas_preprints_copy = DF_perguntas_preprints.copy()

DF_perguntas_pubmed_copy.dropna(subset = ['JIF 2016'], axis = 0, inplace = True)

vetor_JIF_pubmed = np.array(DF_perguntas_pubmed_copy['JIF 2016'])
vetor_score_pubmed = np.array(DF_perguntas_pubmed_copy['Score total'])
```

<h3> Testes estatísticos! </h3>

```python
#Correlação de Spearman

coef_spear, p_value = stats.spearmanr(vetor_JIF_pubmed, vetor_score_pubmed)
print(f'Coeficiente de correlação = {coef_spear};\nValor de p = {p_value}')
ax = sns.scatterplot(x = vetor_JIF_pubmed, y = vetor_score_pubmed)
ax.set_title('JIF 2016 vs. Score Total')
ax.set_ylabel('Score (% of applicable questions)')
ax.set_xlabel('Journal Impact Factor 2016')
sns.set(rc = {'figure.figsize': (10, 5)})
plt.ylim = (0, )
#sns.lmplot(x = vetor_JIF_pubmed, y = vetor_score_pubmed)
```

```python
#Comparação de  pontuações totais entre grupos
vetor_A = DF_perguntas_pubmed['Score total']
vetor_B = DF_perguntas_biorxiv['Score total']
vetor_A_paired = DF_perguntas_peer_rev['Score total']
vetor_B_paired = DF_perguntas_preprints['Score total']

t, p = stats.ttest_ind(vetor_A, vetor_B, alternative = 'two-sided')
t2, p2 = stats.ttest_rel(vetor_A_paired, vetor_B_paired, alternative = 'two-sided')

print(f'>>> Comparação entre grupos individuais - Teste t de Student: t = {t}, p ={p}\n')
print(f'Info sobre pontuações totais do Pubmed' , DF_perguntas_pubmed['Score total'].describe(),'\n')
print(f'Info sobre pontuações totais do biorXiv', DF_perguntas_biorxiv['Score total'].describe(),'\n')
print('___________________________________________________________________________________________\n')
print(f' >>> Comparação pareada - Teste t pareado: t = {t2}, p = {p2}\n')
print(f'Info sobre pontuações totais do Pubmed' , DF_perguntas_peer_rev['Score total'].describe(),'\n')
print(f'Info sobre pontuações totais do biorXiv', DF_perguntas_preprints['Score total'].describe(),'\n')
```
